## Randomness in the generated text

```python
genai.configure(api_key=os.getenv('GOOGLE_API_KEY'))
model=genai.GenerativeModel('gemini-2.5-flash-lite')
prompt='Tell me a good joke on engineer'
response=model.generate_content(prompt,generation_config={'top_k':1})
print(response.text)
```

methods to control randomness in the output:

Temperature(controls the randomness and creativity)
* 0->deteministic,focused and factual
* 1->balanced creativity
* greaterthan 1 more diverse and less coherent

top_P (nucleas sampling-probability cutoff)
top_k(sample from top k likely tokens)
